```python
import os
import pandas as pd
```

```python
i = 0
for file in os.listdir():
    i += 1
    if '.csv' in file:
        print(file)
        df = pd.read_csv(file,skiprows=6)
        df['Date'] = pd.to_datetime(df['Date (MM/DD/YYYY)'])
        df['Hour'] = df['Actual Arrival Time'].apply(lambda x:str(x).split(':')[0])
        df['Hour'] = df['Hour'].apply(lambda x: 0 if str(x)=='24' else x)
        df = df[['Date','Hour']]
        df.dropna(inplace=True)
#         break
#         print(df.head(2))
        if i == 1:
            df.to_csv('../JFK2015.csv',mode='a+',index=False)
        elif i> 1:
            df.to_csv('../JFK2015.csv',mode='a+',header=None,index=False)
```
